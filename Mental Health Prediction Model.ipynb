I've used describe method to find the number of unique value of each features. As we can remember, the Gender has 49 unique value. Is it possible? in today world, it's possible(?). But, for the sake of data analysis, I will make it more simple. 

There are a lot of the unique value that can be merged into one, for example Male, male, m, guy, and other data that lead to male category. For the simplicity purposes, I will just divide the gender on this data into three categories: male, female, and others. 

```python
df['Gender'] = df['Gender'].str.lower()
```

```python
male = re.compile('.*m...*')
female = re.compile('.*f......*')

def gender(x):
    if (female.match(x)) or ('woman' in x) or (x == 'f'):
        return ('female')
    elif (male.match(x)) or ('guy' in x) or (x == 'm'):
        return ('male')
    else:
        return ('others')
```

```python
df['Gender'] = df['Gender'].apply(gender)
df['Gender'].value_counts()
```

```python
America = ['United States', 'Canada', 'Brazil', 'Mexico', 'Colombia', 'Costa Rica', 'Uruguay', 'Bahamas, The']
Europe = ['United Kingdom', 'Germany', 'Ireland', 'Netherlands', 'France', 'Sweden', 'Switzerland', 'Poland',
           'Italy', 'Belgium', 'Bulgaria', 'Austria', 'Finland', 'Croatia', 'Denmark', 'Portugal', 'Greece', 'Norway',
           'Bosnia and Herzegovina', 'Hungary', 'Czech Republic', 'Latvia', 'Romania', 'Spain', 'Georgia',
           'Slovenia', 'Moldova', 'Russia']
Others = ['Australia', 'India', 'New Zealand', 'South Africa', 'Israel', 'Singapore', 'Japan', 'Thailand', 'Nigeria',
           'Philippines', 'Zimbabwe', 'China']

def continent(x):
    if (x in America):
        return 'America'
    elif (x in Europe):
        return 'Europe'
    elif (x in Others):
        return 'Others'
    else:
        print(x)
```

# Clustering (Detecting Outliers)

```python
df_toTrain = pd.get_dummies(df.drop(['Age'], axis=1))
df_toTrain['Age'] = df['Age']
df_toTrain.head()
```

```python
df_toTrain['Age'] = df_toTrain['Age']/100
```

```python
df_toTrain.describe()['Age']
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=88)
```

```python
kmeans.fit(df_toTrain)
```

```python
Y1 = 0
Y0 = 0
N1 = 0
N0 = 0

for i,j in zip(df['treatment'], kmeans.labels_):
    if (i=='Yes'):
        if (j==1):
            Y1+=1
        else:
            Y0+=1
    else:
        if(j==1):
            N1+=1
        else:
            N0+=1
            
print("Clustering results for label 1 and Treatment = Yes are " + str(Y1))
print("Clustering results for label 0 and Treatment = Yes are " + str(Y0))
print("Clustering results for label 1 and Treatment = No are " + str(N1))
print("Clustering results for label 0 and Treatment = No are " + str(N0))
```

```python
minN = min(N1, N0)

if (minN == N1):
    minN = 1
else:
    minN = 0
```

We will delete all of the data(s) that indicate the treatment = No and have the lowest numbers in K-Means label (label 1). Otherwords, we will delete the outliers in treatment = No.

```python
dfNew = []
for i,j,k in zip(df['treatment'], kmeans.labels_, range(len(df['treatment']))):
    if (i=='No' and j!=minN):
        dfNew.append(df.iloc[k,:])
    elif (i=='Yes'):
        dfNew.append(df.iloc[k,:])
```

```python
dfNew = pd.DataFrame(dfNew, columns=df.columns)
```

```python
from sklearn.model_selection import GridSearchCV

def calc_train_error(X_train, y_train, model):
    predictions = model.predict(X_train)
    predictProba = model.predict_proba(X_train)
    matt = matthews_corrcoef(y_train, predictions)
    f1 = f1_score(y_train, predictions, average='macro')
    report = classification_report(y_train, predictions)
    roc_auc = roc_auc_score(y_train, predictProba[:,1])
    acc = accuracy_score(y_train, predictions)
    confMatrix = confusion_matrix(y_train, predictions)
    return {'report': report, 'matthew': matt, 'f1': f1, 'roc': roc_auc, 'accuracy': acc, 'confusion': confMatrix}

def calc_validation_error(X_test, y_test, model):
    predictions = model.predict(X_test)
    predictProba = model.predict_proba(X_test)
    matt = matthews_corrcoef(y_test, predictions)
    f1 = f1_score(y_test, predictions, average='macro')
    report = classification_report(y_test, predictions)
    roc_auc = roc_auc_score(y_test, predictProba[:,1])
    acc = accuracy_score(y_test, predictions)
    confMatrix = confusion_matrix(y_test, predictions)
    return {'report': report, 'matthew': matt, 'f1': f1, 'roc': roc_auc, 'accuracy': acc, 'confusion': confMatrix}

def calc_metrics(X_train, y_train, X_test, y_test, model):
    model.fit(X_train, y_train)
    train_error = calc_train_error(X_train, y_train, model)
    validation_error = calc_validation_error(X_test, y_test, model)
    return train_error, validation_error

from sklearn.model_selection import KFold
K = 5
kf = KFold(n_splits=K, shuffle=True, random_state=42)
```

```python
parameters = {'max_depth': (10,15,20)}
rfc = RandomForestClassifier(random_state=101, n_estimators=300, min_samples_split=2)
dt = GridSearchCV(rfc, parameters, scoring = 'roc_auc', cv=5)
dt.fit(X_train, y_train)
dt.best_estimator_
```

```python
train_errors = []
validation_errors = []

for train_index, val_index in kf.split(df_toTrain, dfNew['treatment']):
    X_train, X_val = df_toTrain.iloc[train_index], df_toTrain.iloc[val_index]
    y_train, y_val = dfNew['treatment'].iloc[train_index], dfNew['treatment'].iloc[val_index]
    
    rfc = RandomForestClassifier(bootstrap=True, class_weight=None, criterion='gini',
            max_depth=10, max_features='auto', max_leaf_nodes=None,
            min_impurity_decrease=0.0, min_impurity_split=None,
            min_samples_leaf=1, min_samples_split=2,
            min_weight_fraction_leaf=0.0, n_estimators=300, n_jobs=None,
            oob_score=False, random_state=101, verbose=0, warm_start=False)
    
    train_error, val_error = calc_metrics(X_train, y_train, X_val, y_val, rfc)
    
    train_errors.append(train_error)
    validation_errors.append(val_error)
    
dfEvalRFC = []
for tr,val in zip(train_errors, validation_errors):
    dfEvalRFC.append([tr['accuracy'], val['accuracy'], tr['roc'], val['roc'], tr['f1'], val['f1'],
                  tr['matthew'], val['matthew']])

dfEvalRFC = pd.DataFrame(dfEvalRFC, columns=['Train Accuracy','Test Accuracy','Train ROC AUC','Test ROC AUC',
                                       'Train F1 Score','Test F1 Score','Train Matthews Corr Coef',
                                       'Test Matthews Corr Coef'])
dfEvalRFC
```
